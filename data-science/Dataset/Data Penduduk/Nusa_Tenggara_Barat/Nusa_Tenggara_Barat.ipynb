```python
import pandas as pd

# total NTB
total_laki = 2832700
total_perempuan = 2813300

# hitung total keseluruhan
total_keseluruhan = (
    total_laki + total_perempuan
)

# dataframe final
df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)

# simpan csv
df_final.to_csv(
    'jumlah_penduduk_nusa_tenggara_barat.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_nusa_tenggara_barat.json',
    orient='records',
    indent=4
)
```

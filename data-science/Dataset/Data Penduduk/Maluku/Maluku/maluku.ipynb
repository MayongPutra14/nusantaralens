```python
import pandas as pd
import json 
```

```python
data = pd.read_csv("Data Maluku Kotor.csv", skiprows=2)
display(data)
display(data.info())
```

```python
df = data.copy()
df.drop(index=0, inplace=True)
df_clean = df[df['Unnamed: 0'].str.contains('Jumlah', na=False, case=False)]
display(df_clean)
```

```python
total_laki = int(pd.to_numeric(df_clean['Laki-laki'].iloc[0]) )
total_perempuan = int(pd.to_numeric(df_clean['Perempuan'].iloc[0]))
total_penduduk = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_penduduk
}])
display(df_final)
```

```python
df_final.to_csv('jumlah_penduduk_maluku.csv', index=False)
df_final.to_json('jumlah_penduduk_maluku.json', orient='records')
```

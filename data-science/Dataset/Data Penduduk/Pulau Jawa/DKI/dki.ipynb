```python
import pandas as pd
import json 
```

```python
data = pd.read_csv('Data DKI Jakarta Kotor.csv', skiprows=2)
display(data)
display(data.info())
```

```python
df = data.copy()
df = df.drop(columns=['Jumlah'])

df_jumlah = df[df['Unnamed: 0'].str.contains('Jumlah', na=False, case=False)]
total_laki = int(pd.to_numeric(df_jumlah['Laki-laki'].iloc[0]))
total_perempuan = int(pd.to_numeric(df_jumlah['Perempuan'].iloc[0]))
total_penduduk = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_penduduk
}])

display(df_final)
```

```python
df_final.to_csv('jumlah_penduduk_dki_jakarta.csv', index=False)
df_final.to_json('jumlah_penduduk_dki_jakarta.json', orient='records')
```

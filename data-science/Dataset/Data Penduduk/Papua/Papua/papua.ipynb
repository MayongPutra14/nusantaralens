```python
import pandas as pd
import json
```

```python
data = pd.read_csv('Data Papua Kotor.csv')
display(data)
display(data.info())
```

```python
df = data.copy()
df = df.drop(columns=['Penduduk (Laki-Laki + Perempuan) (Ribu)'])
display(df)
df_jumlah = df[df['Kelompok Umur'].str.contains('Jumlah', na=False, case=False)]
total_laki = int(pd.to_numeric(df_jumlah['Penduduk (Laki-Laki) (Ribu)'].iloc[0]) * 1000)
total_perempuan = int(pd.to_numeric(df_jumlah['Penduduk (Perempuan) (Ribu)'].iloc[0])*1000)
total_semua = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_semua
}])

display(df_final)
```

```python
df_final.to_csv('jumlah_penduduk_papua.csv', index='False')
df_final.to_json('jumlah_penduduk_papua.json', orient='records')

```

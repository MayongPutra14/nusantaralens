```python
import pandas as pd
import json
```

```python
data = pd.read_csv('Data Papua Barat Kotor.csv', skiprows=2)
display(data)
display(data.info())
```

```python
df_total = data[data['Unnamed: 0'].str.contains('Papua Barat', na=False, case=False)]

total_laki = int(pd.to_numeric(df_total['Laki-laki'].iloc[0]))
total_perempuan = int(pd.to_numeric(df_total['Perempuan'].iloc[0]))
total_keseluruhan = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)
```

```python
df_final.to_csv('jumlah_penduduk_papua_barat.csv', index=False)
df_final.to_json('jumlah_penduduk_papua_barat.json', orient='records', indent=4)
```

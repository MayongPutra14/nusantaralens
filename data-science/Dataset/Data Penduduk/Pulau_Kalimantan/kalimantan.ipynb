```python
import pandas as pd

file_provinsi = [
    'jumlah_penduduk_kalimantan_barat.csv',
    'jumlah_penduduk_kalimantan_tengah.csv',
    'jumlah_penduduk_kalimantan_selatan.csv',
    'jumlah_penduduk_kalimantan_timur.csv',
    'jumlah_penduduk_kalimantan_utara.csv'
]

data_kalimantan = []

total_laki = 0
total_perempuan = 0

for file in file_provinsi:

    df = pd.read_csv(file)

    laki = int(df['laki_laki'].iloc[0])
    perempuan = int(df['perempuan'].iloc[0])

    total_keseluruhan = laki + perempuan

    nama_provinsi = (
        file
        .replace('jumlah_penduduk_', '')
        .replace('.csv', '')
    )

    data_kalimantan.append({
        'provinsi': nama_provinsi,
        'laki_laki': laki,
        'perempuan': perempuan,
        'total_keseluruhan': total_keseluruhan
    })

    total_laki += laki
    total_perempuan += perempuan

total_pulau = total_laki + total_perempuan

data_kalimantan.append({
    'provinsi': 'total_pulau_kalimantan',
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_pulau
})

df_final = pd.DataFrame(data_kalimantan)

display(df_final)

df_final.to_csv(
    'data_penduduk_kalimantan.csv',
    index=False
)

df_final.to_json(
    'data_penduduk_kalimantan.json',
    orient='records',
    indent=4
)
```

```python
import pandas as pd

file_provinsi = [
   'jumlah_penduduk_kalimantan_barat.csv',
    'jumlah_penduduk_kalimantan_tengah.csv',
    'jumlah_penduduk_kalimantan_selatan.csv',
    'jumlah_penduduk_kalimantan_timur.csv',
    'jumlah_penduduk_kalimantan_utara.csv'
]


total_laki_pulau = 0
total_perempuan_pulau = 0


for filename in file_provinsi:
    df_temp = pd.read_csv(filename)

    total_laki_pulau += int(
        df_temp['laki_laki'].iloc[0]
    )

    total_perempuan_pulau += int(
        df_temp['perempuan'].iloc[0]
    )


total_keseluruhan_pulau = (
    total_laki_pulau + total_perempuan_pulau
)

# dataframe final pulau
df_final_pulau = pd.DataFrame([{
    'keterangan': 'total_pulau_kalimantan',
    'laki_laki': total_laki_pulau,
    'perempuan': total_perempuan_pulau,
    'total_keseluruhan': total_keseluruhan_pulau
}])

display(df_final_pulau)


df_final_pulau.to_json(
    'data_gabungan_kalimantan.json',
    orient='records',
    indent=4
)


df_final_pulau.to_csv(
    'data_gabungan_kalimantan.csv',
    index=False
)
```

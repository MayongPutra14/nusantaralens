```python
import pandas as pd

# total Kalimantan Selatan 2025
total_laki = 2185000
total_perempuan = 2138330

# hitung total keseluruhan
total_keseluruhan = (
    total_laki + total_perempuan
)

# dataframe final
df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)

# simpan csv
df_final.to_csv(
    'jumlah_penduduk_kalimantan_selatan.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_kalimantan_selatan.json',
    orient='records',
    indent=4
)
```

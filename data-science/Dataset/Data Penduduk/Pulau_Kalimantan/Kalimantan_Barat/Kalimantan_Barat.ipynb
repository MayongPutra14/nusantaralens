```python
import pandas as pd


data = pd.read_csv(
    "Data_Kalimantan_Barat_Kotor.csv",
    sep=';'
)

display(data.tail())

total_laki = 2957074
total_perempuan = 2808956

total_keseluruhan = (
    total_laki + total_perempuan
)

# dataframe final
df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)

# simpan csv
df_final.to_csv(
    'jumlah_penduduk_kalimantan_barat.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_kalimantan_barat.json',
    orient='records',
    indent=4
)
```

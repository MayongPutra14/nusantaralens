```python
import pandas as pd


data = pd.read_csv(
    "Data_Kalimantan_Utara_Gabungan.csv"
)

display(data)

total_laki = int(data.iloc[5,1])

total_perempuan = int(data.iloc[5,2])


total_keseluruhan = (
    total_laki + total_perempuan
)


df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)

# simpan csv final
df_final.to_csv(
    'jumlah_penduduk_kalimantan_utara.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_kalimantan_utara.json',
    orient='records',
    indent=4
)
```

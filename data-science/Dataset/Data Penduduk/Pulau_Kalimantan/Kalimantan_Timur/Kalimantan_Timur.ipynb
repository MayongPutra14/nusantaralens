```python
import pandas as pd

laki = pd.read_csv(
    "Data_Kalimantan_Timur_Laki_laki_Kotor.csv",
    sep=','
)


perempuan = pd.read_csv(
    "Data_Kalimantan_Timur_Perempuan_Kotor.csv",
    sep=','
)

display(laki)
display(perempuan)


df_gabung = pd.DataFrame({
    'wilayah': laki.iloc[:,0],
    'laki_laki': laki.iloc[:,1],
    'perempuan': perempuan.iloc[:,1]
})

display(df_gabung)


df_gabung.to_csv(
    'Data_Kalimantan_Timur_Gabungan.csv',
    index=False
)
```

```python
import pandas as pd


total_laki = int(df_gabung.iloc[17,1])

total_perempuan = int(df_gabung.iloc[17,2])


total_keseluruhan = (
    total_laki + total_perempuan
)

# dataframe final
df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)


df_final.to_csv(
    'jumlah_penduduk_kalimantan_timur.csv',
    index=False
)


df_final.to_json(
    'jumlah_penduduk_kalimantan_timur.json',
    orient='records',
    indent=4
)
```

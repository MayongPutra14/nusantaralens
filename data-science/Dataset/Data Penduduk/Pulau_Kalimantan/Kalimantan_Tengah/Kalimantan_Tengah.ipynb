```python
import pandas as pd

data = pd.read_csv(
    "Data_Kalimantan_Tengah_Kotor.csv",
    sep=';'
)

display(data.tail())


total_laki = 1450000
total_perempuan = 1370000


total_keseluruhan = (
    total_laki + total_perempuan
)


df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)

# simpan csv
df_final.to_csv(
    'jumlah_penduduk_kalimantan_tengah.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_kalimantan_tengah.json',
    orient='records',
    indent=4
)
```

```python
import pandas as pd


total_laki = 2871257
total_perempuan = 2871307


total_keseluruhan = (
    total_laki + total_perempuan
)


df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)


df_final.to_csv(
    'jumlah_penduduk_nusa_tenggara_timur.csv',
    index=False
)


df_final.to_json(
    'jumlah_penduduk_ntt.json',
    orient='records',
    indent=4
)
```

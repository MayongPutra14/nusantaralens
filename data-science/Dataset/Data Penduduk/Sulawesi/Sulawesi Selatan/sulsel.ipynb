```python
import pandas as pd
import json
```

```python
data = pd.read_csv('Data Sulawesi Selatan Kotor.csv')
display(data)
display(data.info())
```

```python
df = data.copy()
df = df.drop(columns=['Penduduk (Laki-Laki + Perempuan) (Ribu)'])
df_jumlah = df[df['Kelompok Umur'].str.contains('Jumlah/Total', na=False, case=False)]
total_laki = int(pd.to_numeric(df_jumlah['Penduduk (Laki-Laki) (Ribu)'].iloc[0])*1000)
total_perempuan = int(pd.to_numeric(df_jumlah['Penduduk (Perempuan) (Ribu)'].iloc[0])*1000)
total_penduduk = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_penduduk
}])

display(df_final)
```

```python
df_final.to_csv('jumlah_penduduk_sulawesi_selatan.csv', index=False)
df_final.to_json('jumlah_penduduk_sulawesi_selatan.json', orient='records')
```

```python
import pandas as pd

data = pd.read_csv("Data_Aceh_Kotor.csv")

df = data.copy()

df_jumlah = df[
    df.iloc[:,0].astype(str).str.contains(
        'PROVINSI ACEH',
        na=False,
        case=False
    )
]

total_laki = int(
    pd.to_numeric(
        df_jumlah.iloc[0,1],
        errors='coerce'
    )
)

total_perempuan = int(
    pd.to_numeric(
        df_jumlah.iloc[0,2],
        errors='coerce'
    )
)


total_penduduk = total_laki + total_perempuan

# dataframe final
df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_penduduk
}])

display(df_final)

# simpan csv
df_final.to_csv(
    'jumlah_penduduk_aceh.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_aceh.json',
    orient='records',
    indent=4
)
```

```python
import pandas as pd

data = pd.read_csv("Data_Kepulauan_Riau_Kotor.csv")

df = data.copy()

df_jumlah = df[
    df.iloc[:,0].astype(str).str.contains(
        'Kepulauan Riau',
        na=False,
        case=False
    )
]

total_laki = int(
    float(df_jumlah.iloc[0,1]) * 1000
)

total_perempuan = int(
    float(df_jumlah.iloc[0,2]) * 1000
)

total_penduduk = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_penduduk
}])

display(df_final)

# simpan csv
df_final.to_csv(
    'kepulauan_riau.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_kepulauan_riau.json',
    orient='records',
    indent=4
)
```

```python
import pandas as pd


data = pd.read_csv("Data_Sumatera_Selatan_Kotor.csv")

df = data.copy()

total_laki = int(df.iloc[2,1])

total_perempuan = int(df.iloc[2,2])

total_keseluruhan = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)

# simpan csv
df_final.to_csv(
    'jumlah_penduduk_sumatera_selatan.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_sumatera_selatan.json',
    orient='records',
    indent=4
)
```

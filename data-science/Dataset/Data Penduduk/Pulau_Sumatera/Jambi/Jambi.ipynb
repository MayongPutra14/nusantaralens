```python
import pandas as pd

data = pd.read_csv(
    "Data_Jambi_Kotor.csv",
    sep=';'
)

df = data.copy()

df.columns = [
    'kelompok_umur',
    'laki_laki',
    'perempuan'
]

display(df.head())

df_jumlah = df[
    df['kelompok_umur'].astype(str).str.contains(
        'Jumlah/Total',
        na=False
    )
]

total_laki = int(
    float(df_jumlah.iloc[0,1]) * 1000
)

total_perempuan = int(
    float(df_jumlah.iloc[0,2]) * 1000
)

total_keseluruhan = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)

df_final.to_csv(
    'jumlah_penduduk_jambi.csv',
    index=False
)

df_final.to_json(
    'jumlah_penduduk_jambi.json',
    orient='records',
    indent=4
)
```

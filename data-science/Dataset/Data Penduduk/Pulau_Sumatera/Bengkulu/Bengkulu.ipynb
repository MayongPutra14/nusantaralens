```python
import pandas as pd

# total penduduk Bengkulu 2025
total_laki = 1091256
total_perempuan = 1046788

total_keseluruhan = (
    total_laki + total_perempuan
)

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_keseluruhan
}])

display(df_final)

# simpan csv
df_final.to_csv(
    'jumlah_penduduk_bengkulu.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_bengkulu.json',
    orient='records',
    indent=4
)
```

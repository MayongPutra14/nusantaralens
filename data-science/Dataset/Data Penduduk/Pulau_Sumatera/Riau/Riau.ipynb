```python
import pandas as pd

data = pd.read_csv("Data_Riau_Kotor.csv")

df = data.copy()

baris_terakhir = df.tail(1)

total_laki = int(
    pd.to_numeric(
        baris_terakhir.iloc[0,1],
        errors='coerce'
    )
)

total_perempuan = int(
    pd.to_numeric(
        baris_terakhir.iloc[0,2],
        errors='coerce'
    )
)

total_penduduk = total_laki + total_perempuan

df_final = pd.DataFrame([{
    'laki_laki': total_laki,
    'perempuan': total_perempuan,
    'total_keseluruhan': total_penduduk
}])

display(df_final)
```

```python
# simpan csv
df_final.to_csv(
    'jumlah_penduduk_riau.csv',
    index=False
)

# simpan json
df_final.to_json(
    'jumlah_penduduk_riau.json',
    orient='records',
    indent=4
)
```

# DATA WRANGLING 

## **LOAD DATA**

```python
df = pd.read_csv("kamus_perbandingan_nusantaralens (1).csv")
display(df)
```

```python
print("Mengecek tipe data kamus")
display(df.info())
```

```python
df_cleaned = df.apply(lambda x: x.str.strip() if x.dtype=='object' else x)

df_cleaned = df_cleaned.replace(';' , ',', regex=True)
display(df_cleaned)
```

```python
display(df_cleaned.describe())

display(df_cleaned.isna().sum())

display(df_cleaned.duplicated().sum())
```

```python
columns = ['Aceh','Jawa', 'Abui']

for col in columns:
    kamus_pisah = df_cleaned[['Indonesia', col]]
    kamus_pisah = kamus_pisah.dropna()
    file = f'{col}.csv'
    kamus_pisah.to_csv(file, index=False)
    print(f'Berhasil Menyimpan {file}')
```

```python
tampung_data = {}
daftar_file = ['Aceh', 'Jawa', 'Abui']

for tabel in daftar_file:
    df = f'{tabel}.csv'
    tampung_data[tabel] = pd.read_csv(df)
    print(f'Berhasil Load tabel dari {df}')
```

```python
for nama, df in tampung_data.items():
    print(f'Tabel Bahasa {nama}')
    print(f'Jumlah baris kolom {df.shape}' )
    print(f'Informasi tabel {nama} adalah {df.info()}')
```
